```python
class Hospital:
    def __init__(self):
        self.filename = 'Patients.csv'

    # Adding Patients
    def add_patient(self):
        name = input('Enter the name: ')
        age = input('Enter the age: ')
        disease = input('Enter the Disease: ')

        file = open(self.filename, 'a')
        file.write(f'{name}, {age}, {disease} \n')
        file.close()

        print('Patient Added Successfully')

    # Viewing Patient
    def view_patients(self):
        try:
            file = open(self.filename, 'r')
            data = file.readlines()
            file.close()

            if len(data) == 0:
                print('No Record Added')
                return
            print('\nPatient records:')
            for i in data:
                name, age, disease = i.strip().split(',')
                print(f'Name:{name} |Age:{age} |Disease:{disease}')
        except:
            print('file not found')

    # Search Patient
    def seatch_patient(self):
        search_name = input('Enter the Patient name to search: ')

        try:
            file = open(self.filename, 'r')
            data = file.readlines()
            file.close()

            for i in data:
                name, age, disease = i.strip().split(',')
                if name.lower() == search_name.lower():
                    print(f'Found: {name}, {age}, {disease}')
                    found = True

            if not found:
                print('Patient Not Found')

        except:
            print('No file found')

    # deleting Patient from data
    def delete_patient(self):
        delete_name = input('enter patient name to delete: ')

        try:
            file = open(self.filename, 'r')
            data = file.readlines()
            file.close()

            file = open(self.filename, 'w')
            found = False

            for i in data:
                name, age, disease = i.strip().split(',')
                if name.lower() != delete_name.lower():
                    file.write(i)
                else:
                    found = True

            file.close()

            if found:
                print('Patient Delete Successfully!')
            else:
                print('Patient not found.')
        except:
            print('No file Found')

    # Updating Patient details in data
    def update_patient(self):
        update_name = input('Enter patient name: ')

        try:
            file = open(self.filename, 'r')
            data = file.readlines()
            file.close()

            file = open(self.filename, 'w')
            found = False

            for i in data:
                name, age, disease = i.strip().split(',')
                if name.lower() == update_name.lower():
                    new_age = input('Enter new age: ')
                    new_disease = input('Enter new disease: ')
                    file.write(f'{name}, {new_age}, {new_disease}\n')
                    found = True
                else:
                    file.write(i)
            file.close()

            if found:
                print('Patient Update Successfully!')
            else:
                print('Patient Not found')

        except:
            print('No File found')

def main():
    hospital = Hospital()

    while True:
        print('\n Hospital Management System')
        print('1. Add Patient')
        print('2. View Patient')
        print('3. Search Patient')
        print('4. Delete Patient')
        print('5. Update Patient')
        print('6. Exit')

        
        choice = input('Enter Your Choice: ')

        if choice == '1':
            hospital.add_patient()
        elif choice == '2':
            hospital.view_patients()
        elif choice == '3':
            hospital.seatch_patient()
        elif choice == '4':
            hospital.delete_patient()
        elif choice == '5':
            hospital.update_patient()
        elif choice =='6':
            print('Program Closed')
            break
        else:
            print('Invalid choice')
```

```python
main()
```
